```python
# sphinx_gallery_thumbnail_number = 11

import matplotlib.pyplot as plt
import numpy as np

# Some example data to display
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)
```

```python
fig, axs = plt.subplots(2, 2)
axs[0, 0].plot(x, y)
axs[0, 0].set_title('Axis [0,0]')
axs[0, 1].plot(x, y, 'tab:orange')
axs[0, 1].set_title('Axis [0,1]')
axs[1, 0].plot(x, -y, 'tab:green')
axs[1, 0].set_title('Axis [1,0]')
axs[1, 1].plot(x, -y, 'tab:red')
axs[1, 1].set_title('Axis [1,1]')

for ax in axs.flat:
    ax.set(xlabel='x-label', ylabel='y-label')

# Hide x labels and tick labels for top plots and y ticks for right plots.
for ax in axs.flat:
    ax.label_outer()
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, sharex='all', sharey='all')
fig.suptitle('Sharing x per column, y per row')
ax1.plot(x, y)
ax2.plot(x, y**2, 'tab:orange')
ax3.plot(x, -y, 'tab:green')
ax4.plot(x, -y**2, 'tab:red')

ax = fig.add_subplot(111, frameon=False)
ax.tick_params(labelcolor='none', top='off', bottom='off', left='off', right='off')
plt.xlabel("common X")
plt.ylabel("common Y")

for ax in fig.get_axes():
    ax.label_outer()
```

```python
fig, axs = plt.subplots(2, 2, sharex='all', sharey='all', figsize=(12,9),
                        gridspec_kw={'hspace': 0, 'wspace': 0})
(ax1, ax2), (ax3, ax4) = axs
fig.suptitle('Sharing x per column, y per row')
ax1.set_xlim(0.8,5.2)
ax1.plot(x, y)
ax2.plot(x, y**2, 'tab:orange')
ax3.plot(x, -y, 'tab:green')
ax4.plot(x, -y**2, 'tab:red')

props = dict(boxstyle='round', facecolor='silver', alpha=0.3)
ax1.set_title('Axis [0,0]', pad=-30, fontsize='xx-large', bbox=props)
ax2.set_title('Axis [0,1]', pad=-30, fontsize='xx-large', bbox=props)
ax3.set_title('Axis [1,0]', pad=-30, fontsize='xx-large')
ax4.set_title('Axis [1,1]', pad=-30, fontsize='xx-large')

#ax = fig.add_subplot(111, frameon=False)
#ax.tick_params(labelcolor='none', top='off', bottom='off', left='off', right='off')
#plt.xlabel("common X1")
#plt.ylabel("common Y")
#ax.yaxis.set_label_coords(-0.1, 0.5)

fig.text(0.5, 0.06, 'x-label', ha='center', fontsize='x-large')
fig.text(0.05, 0.5, 'y-label', va='center', rotation='vertical', fontsize='x-large')

#for ax in axs.flat:
#    ax.set(xlabel='x-label', ylabel='y-label')
    
#for ax in fig.get_axes():
#    ax.label_outer()
```
